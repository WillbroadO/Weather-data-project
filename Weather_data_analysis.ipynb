```python
api_key = "13e8c25b74a11a570eb63c369d7b6f9e"
lat = "-0.537544"
lon = "34.126236"
url = "https://api.openweathermap.org/data/2.5/weather?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)

response = requests.get(url)
data_oremo = json.loads(response.text)
data_oremo = pd.json_normalize(data_oremo)

data_oremo['Concession'] = 'oremo'

lat = "-0.553724"
lon = "34.092075"
url = "https://api.openweathermap.org/data/2.5/weather?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)

response = requests.get(url)
data_gobi = json.loads(response.text)
data_gobi = pd.json_normalize(data_gobi)

data_gobi['Concession'] = 'gobi'

lat = "-0.546308"
lon = "34.100975"
url = "https://api.openweathermap.org/data/2.5/weather?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)

response = requests.get(url)
data_ant = json.loads(response.text)
data_ant = pd.json_normalize(data_ant)

data_ant['Concession'] = 'ant'

lat = "-0.539299"
lon = "34.105053"
url = "https://api.openweathermap.org/data/2.5/weather?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)

response = requests.get(url)
data_kal = json.loads(response.text)
data_kal = pd.json_normalize(data_kal)

data_kal['Concession'] = 'kal'
```

Combing the four datasets

```python
weather_df = pd.concat([data_oremo, data_gobi, data_ant, data_kal], axis=0)
weather_df['Date'] = pd.to_datetime('today')
weather_df
```

Exporting the data

```python
weather_df.to_excel('original.xlsx')
```

Loading the exported file back

```python
old_file = pd.read_excel('original.xlsx')
```

```python
new_df = pd.concat([weather_df, old_file], axis=0)
new_df
```
